```python
import math 
import random
```

```python
def distancia(A,B):
    if type(A) is int or type(A) is float:
        return abs(B-A)
    else:
        return math.sqrt(sum([(A[i]-B[i])**2 for i in range(len(A))]))
```

```python
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = float(time() - inicio)
        print("\r\n Tiempo de ejecución para algoritmo: "+"{0:.25f}".format(tiempo))
        return resultado
    
    return wrapper
```

### DIVIDE Y VENCERÁS

```python
def distancia_fuerza_bruta(L):
    mejor_distancia = 100000e10
    
    A,B=(),()
    
    for i in range(len(L)):
        for j in range(i+1, len(L)):
            if distancia(L[i],L[j]) < mejor_distancia:
                A,B=L[i],L[j]
                mejor_distancia = distancia(L[i],L[j])
    return [A,B] 

def distancia_divide_y_venceras(L):
    #si hay pocos por fuerza bruta
    if len(L) < 10:
        return distancia_fuerza_bruta(L)
    else:
        #dividir en listas grandes
        L_sorted =  sorted(L, key=lambda x:x[0])#Ordenamos la lista grande y luego dividimos
        
        Lista_IZDA=L_sorted[:len(L)//2]
        Lista_DCHA=L_sorted[len(L)//2:]
        
        puntos_lista_izda=distancia_divide_y_venceras(Lista_IZDA)
        puntos_lista_dcha=distancia_divide_y_venceras(Lista_DCHA)
        
        return distancia_fuerza_bruta(puntos_lista_izda + puntos_lista_dcha)

@calcular_tiempo
def LANZA(L):
    return distancia_divide_y_venceras(L)

SOL= LANZA(LISTA_2D[:100])
print(SOL)
```

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

def precios(tarifas):
    N=len(TARIFAS[0])
    
    PRECIOS = [[9999]*N for i in [9999]*N]
    RUTAS = [[""]*N for i in [9999]*N]
    
    for i in range(N-1):
        for j in range(i+1,N):
            MIN=TARIFAS[i][j]
            RUTAS[i][j]=i
            
            for k in range (i,j):
                if PRECIOS[i][k] + TARIFAS [k][j] < MIN:
                    MIN = min(MIN, PRECIOS [i][k]+ TARIFAS[k][j])
                    RUTAS[i][j]=k
            PRECIOS [i][j]= MIN
            
    return PRECIOS, RUTAS
                
                
PRECIOS, RUTAS = precios(TARIFAS) 
print(PRECIOS)
print()             
print(RUTAS)

def calcular_ruta(RUTA, desde, hasta):
    if desde == hasta:
        #print("Ir a :" + str(desde))
        return desde 
    else:
        return str(calcular_ruta(RUTA, desde, RUTA[desde][hasta])) +  ',' + str(RUTA[desde][hasta]) 

print("\nLa ruta es:")  
calcular_ruta(RUTAS, 0,6)
```
